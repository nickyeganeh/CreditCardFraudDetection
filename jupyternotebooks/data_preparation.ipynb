```python
def treat_outliers(data, var):
    df = data.copy()

    for i in var:
        outlier_variable = detect_outliers(df[i])
        df.loc[outlier_variable,i] = np.median(df[i])
        
    return df

def detect_outliers(data):
    outliers = []
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    IQR = q3 - q1
    lower_bound = q1 - (IQR * 1.5)
    upper_bound = q3 + (IQR * 1.5)
    
    for i,j in enumerate(data):
        if j < lower_bound or j > upper_bound:
            outliers.append(i)
            
    return outliers
```

```python
var = list(df.select_dtypes(include=['float64']).columns)
df1 = treat_outliers(df, var)
```

```python
plt.figure(figsize = (18, 18))
for i, col in enumerate(df1.select_dtypes(include=['float64']).columns):
    ax = plt.subplot(6, 5, i + 1)
    sns.histplot(data = df1, x = col, ax = ax, kde = True)
plt.suptitle('Data distribution of continuous variables')
plt.tight_layout()
plt.show();
```

```python
plt.figure(figsize = (18, 18))
for i, col in enumerate(df1.select_dtypes(include = ['float64']).columns):
    ax = plt.subplot(6, 5, i + 1)
    sns.boxplot(data = df1, x = col, ax = ax)
plt.suptitle('Box Plot of continuous variables')
plt.tight_layout()
plt.show();
```

```python
from sklearn.feature_selection import f_classif

X = df.iloc[:,0:-1]
y = df.iloc[:,-1]

fs = f_classif(X, y)

selected_features = []
for i, j in enumerate(X.columns):
    if fs[1][i] >= 0.05:
        selected_features.append(j)

X = X[selected_features]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
pd.DataFrame(X_train).to_csv("../train-test_data/X_train.csv")
pd.DataFrame(X_test).to_csv("../train-test_data/X_test.csv")
pd.DataFrame(y_train).to_csv("../train-test_data/y_train.csv")
pd.DataFrame(y_test).to_csv("../train-test_data/y_test.csv")
```
